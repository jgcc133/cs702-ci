```python
import copy
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

from scipy import stats

plt.style.use("seaborn-v0_8")
```

```python
# Define the environment to generate the synthetic data
class Environment:
    params = [[7., 5.],  # Parameters for Person A
              [4., 8.]]  # Parameters for Person B
    
    @classmethod
    def sample(cls, person: int, task: int, size=1):
        return stats.gamma.rvs(
            cls.params[person][task],
            loc=1, 
            scale=1, 
            size=size
        )

p00 = Environment.sample(0, 0, size=25)
p01 = Environment.sample(0, 1, size=25)
p10 = Environment.sample(1, 0, size=25)
p11 = Environment.sample(1, 1, size=25)

# Plot historgrams of the samples from each task
# Set the range to be the same for all plots
fig, ax = plt.subplots(4, 1)
ax[0].hist(p00, bins=20)
ax[1].hist(p01, bins=20)
ax[2].hist(p10, bins=20)
ax[3].hist(p11, bins=20)
ax[0].set(xlim=(0, 20), title='Person A, Task 0')
ax[1].set(xlim=(0, 20), title='Person A, Task 1')
ax[2].set(xlim=(0, 20), title='Person B, Task 0')
ax[3].set(xlim=(0, 20), title='Person B, Task 1')


plt.tight_layout()
plt.show()
```

```python
person_idx = np.hstack([
    np.repeat(0, 50), 
    np.repeat(1, 50)
])
task_idx = np.hstack([
    np.repeat(0, 25), 
    np.repeat(1, 25), 
    np.repeat(0, 25), 
    np.repeat(1, 25)
])
p = np.hstack([p00, p01, p10, p11])

df = pd.DataFrame({
    'person': person_idx,
    'task': task_idx,
    'p': p
})
# df.to_csv('data.csv', index=False)
print(df)
```
